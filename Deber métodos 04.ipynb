â€¢	[0,1]
â€¢	[1, 3.2] 
â€¢	[3.2, 4]


```python
def f(x):
    return x**3 - 7*x**2 + 14*x - 6

def biseccion(a, b, tol):
    while (b - a) / 2 > tol:
        c = (a + b) / 2
        if f(c) == 0:
            break
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2

tolerancia = 1e-2

# Intervalos dados
intervalos = [(0, 1), (1, 3.2), (3.2, 4)]

# Soluciones
soluciones = []

for intervalo in intervalos:
    a, b = intervalo
    solucion = biseccion(a, b, tolerancia)
    soluciones.append(solucion)
    print(f"La soluciÃ³n en el intervalo [{a}, {b}] es: {solucion:.3f}")


```

con radio ð‘Ÿ. (Consulte la figura adjunta.) Cuando se llena con agua hasta una 
distancia â„Ž a partir de la parte superior, el volumen ð‘‰ de agua es.
V = L[0.5 Ï€r^2 âˆ’ r^2arsen (h/r) âˆ’ h(r^2 âˆ’ h^2)^1/2

Suponga que ð‘³ = ðŸðŸŽ ð’„ð’Ž, ð’“ = ðŸð’„ð’Ž ð’š ð‘½ = ðŸðŸ. ðŸ’ ð’„ð’ŽðŸ‘
. Encuentre la profundidad del agua en el abrevadero dentro de 0.01 ð‘ð‘š

```python
import math

# ParÃ¡metros dados
L = 10  # Longitud del abrevadero en cm
r = 1   # Radio del semicÃ­rculo en cm
V = 12.43  # Volumen de agua en cm^3

# FunciÃ³n para el cÃ¡lculo de h
def calculate_h(h):
    return math.sqrt(r**2 - ((0.5 * math.pi * r**2 - r**2 * math.asin(h/r) - V/L) / h)**2) - h

# MÃ©todo de Newton-Raphson para encontrar la raÃ­z
def newton_raphson(guess, tolerance=0.01):
    while True:
        h_next = guess - calculate_h(guess) / (1 + ((r**2 - guess**2) / guess**2))
        if abs(h_next - guess) < tolerance:
            return h_next
        guess = h_next

# AproximaciÃ³n inicial
initial_guess = 0.5

# Calcula la profundidad del agua
water_depth = newton_raphson(initial_guess)

print("La profundidad del agua en el abrevadero es:", round(water_depth, 2), "cm")
```

Un objeto que cae verticalmente a travÃ©s del aire estÃ¡ sujeto a una resistencia 
viscosa, asÃ­ como a la fuerza de gravedad. Suponga que un objeto con masa ð‘š cae 
desde una altura ð’”ðŸŽ y que la altura del objeto despuÃ©s de ð‘¡ segundos es
ð‘ (ð‘¡) = ð‘ 0 âˆ’(ð‘šð‘”/ð‘˜)ð‘¡ +ð‘š^2ð‘”/ð‘˜^2(1 âˆ’ ð‘’^(âˆ’ð‘˜ð‘¡/ð‘š) )

donde ð‘” = 9.81 ð’Ž/ð’”^ðŸ y ð‘˜ representa el coeficiente de la resistencia del aire en ð‘µð’”/ð’Ž. 
Suponga ð’”ðŸŽ = 300 ð‘š, ð‘š = 0.25 ð‘˜ð‘” y ð‘˜ = 0.1 ð‘µð’”/ð’Ž. Encuentre, dentro de 0.01 ð‘ ð‘’ð‘”ð‘¢ð‘›ð‘‘ð‘œð‘ , 
el tiempo que tarda un cuarto de kg en golpear el piso.

```python
import math

# ParÃ¡metros dados
s0 = 300  # Altura inicial en metros
m = 0.25  # Masa del objeto en kg
g = 9.81  # AceleraciÃ³n debido a la gravedad en m/s^2
k = 0.1   # Coeficiente de resistencia del aire en Ns/m

# FunciÃ³n para calcular s(t)
def calculate_s(t):
    return s0 - (m * g / k) * t + (m**2 * g / k**2) * (1 - math.exp(-k * t / m))

# Derivada de la funciÃ³n s(t)
def derivative_s(t):
    return -(m * g / k) + (m**2 * g / k**2) * (k / m) * math.exp(-k * t / m)

# MÃ©todo de Newton-Raphson para encontrar la raÃ­z
def newton_raphson(guess, tolerance=0.01):
    while True:
        t_next = guess - calculate_s(guess) / derivative_s(guess)
        if abs(t_next - guess) < tolerance:
            return t_next
        guess = t_next

# AproximaciÃ³n inicial
initial_guess = 1

# Calcula el tiempo que tarda en golpear el piso
time_to_hit_ground = newton_raphson(initial_guess)

print("El tiempo que tarda en golpear el piso es:", round(time_to_hit_ground, 2), "segundos")
```

Use el teorema 2.1 para encontrar una cota para el nÃºmero de iteraciones necesarias 
para lograr una aproximaciÃ³n con precisiÃ³n de ðŸðŸŽ^âˆ’ðŸ’ para la soluciÃ³n de ð’™^ðŸ‘ âˆ’ ð’™ âˆ’ðŸ = ðŸŽ que se encuentra dentro del intervalo [1, 2]. Encuentre una aproximaciÃ³n para la raÃ­z con este grado de precisiÃ³n.


```python
def f(x):
    return x**3 - x - 1

def metodo_biseccion(a, b, tol, max_iteraciones):
    for i in range(1, max_iteraciones + 1):
        c = (a + b) / 2
        if abs(f(c)) < tol:
            return c, i
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2, max_iteraciones

# Definir los parÃ¡metros
a = 1
b = 2
tol = 1e-4
max_iteraciones = 100 

# LLamamos a la funciÃ³n de mÃ©todo de bisecciÃ³n
raiz, iteraciones = metodo_biseccion(a, b, tol, max_iteraciones)

# Mostrar el resultado
print("AproximaciÃ³n de la raÃ­z:", raiz)
print("NÃºmero de iteraciones:", iteraciones)
```
