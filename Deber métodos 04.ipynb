Use el método de bisección para encontrar soluciones precisas dentro de 𝟏𝟎 ^−𝟐 para 𝒙^𝟑 – 𝟕𝒙^𝟐 + 𝟏𝟒𝒙 − 𝟔 = 𝟎 en cada intervalo.
•	[0,1]
•	[1, 3.2] 
•	[3.2, 4]


```python
def f(x):
    return x**3 - 7*x**2 + 14*x - 6

def biseccion(a, b, tol):
    while (b - a) / 2 > tol:
        c = (a + b) / 2
        if f(c) == 0:
            break
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2

tolerancia = 1e-2

# Intervalos dados
intervalos = [(0, 1), (1, 3.2), (3.2, 4)]

# Soluciones
soluciones = []

for intervalo in intervalos:
    a, b = intervalo
    solucion = biseccion(a, b, tolerancia)
    soluciones.append(solucion)
    print(f"La solución en el intervalo [{a}, {b}] es: {solucion:.3f}")


```

Use el método de bisección para encontrar una aproximación dentro de 10^-3 para 
un valor en [−2, 0] con 𝑥^2 − 1 = ℯ^(1-x^2).

```python
import numpy as np

# Definimos la función f(x) = x^2 - 1 - e^(1 - x^2)
def f(x):
    return x**2 - 1 - np.exp(1 - x**2)

# Función de bisección
def biseccion(a, b, tol):
    while (b - a) / 2 > tol:
        c = (a + b) / 2
        if f(c) == 0:
            return c
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2

# Tolerancia
tolerancia = 1e-3

# Definimos el intervalo inicial donde se sospecha que está la raíz
a = -2
b = 0

# Encontrar la solución utilizando el método de bisección
solucion = biseccion(a, b, tolerancia)
print(f"La solución es: {solucion:.5f}")
```

 Sea 𝑓(𝑥) = (𝑥 + 3)*((𝑥 + 1)^2)*𝑥((𝑥 − 1)^3)(𝑥 − 3). ¿En qué cero de 𝑓 converge el método 
de bisección cuando se aplica en los siguientes intervalos?
a. [−1.5, 2.5] b. [−0.5, 2.4] c. [−0.5, 3] d. [−3, −0.5]

```python
def f(x):
    return (x + 3) * ((x + 1) ** 2) * x * ((x - 1) ** 3) * (x - 3)

def bisection_method(f, a, b, tol=1e-6, max_iter=1000):
    if f(a) * f(b) > 0:
        print("No se puede garantizar la convergencia en el intervalo dado.")
        return None

    iter_count = 0
    while iter_count < max_iter:
        c = (a + b) / 2
        if abs(f(c)) < tol:
            return c
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
        iter_count += 1

    print("El método de la bisección no converge después de {} iteraciones.".format(max_iter))
    return None

# Intervalo a: [-1.5, 2.5]
print("Intervalo a: [-1.5, 2.5]")
print("Cero de la función donde converge el método de la bisección:", bisection_method(f, -1.5, 2.5))

# Intervalo b: [-0.5, 2.4]
print("\nIntervalo b: [-0.5, 2.4]")
print("Cero de la función donde converge el método de la bisección:", bisection_method(f, -0.5, 2.4))

# Intervalo c: [-0.5, 3]
print("\nIntervalo c: [-0.5, 3]")
print("Cero de la función donde converge el método de la bisección:", bisection_method(f, -0.5, 3))

# Intervalo d: [-3, -0.5]
print("\nIntervalo d: [-3, -0.5]")
print("Cero de la función donde converge el método de la bisección:", bisection_method(f, -3, -0.5))
```

Un abrevadero de longitud 𝐿 tiene una sección transversal en forma de semicírculo 
con radio 𝑟. (Consulte la figura adjunta.) Cuando se llena con agua hasta una 
distancia ℎ a partir de la parte superior, el volumen 𝑉 de agua es.
V = L[0.5 πr^2 − r^2arsen (h/r) − h(r^2 − h^2)^1/2

Suponga que 𝑳 = 𝟏𝟎 𝒄𝒎, 𝒓 = 𝟏𝒄𝒎 𝒚 𝑽 = 𝟏𝟐. 𝟒 𝒄𝒎𝟑
. Encuentre la profundidad del agua en el abrevadero dentro de 0.01 𝑐𝑚

```python
import math

# Parámetros dados
L = 10  # Longitud del abrevadero en cm
r = 1   # Radio del semicírculo en cm
V = 12.43  # Volumen de agua en cm^3

# Función para el cálculo de h
def calculate_h(h):
    return math.sqrt(r**2 - ((0.5 * math.pi * r**2 - r**2 * math.asin(h/r) - V/L) / h)**2) - h

# Método de Newton-Raphson para encontrar la raíz
def newton_raphson(guess, tolerance=0.01):
    while True:
        h_next = guess - calculate_h(guess) / (1 + ((r**2 - guess**2) / guess**2))
        if abs(h_next - guess) < tolerance:
            return h_next
        guess = h_next

# Aproximación inicial
initial_guess = 0.5

# Calcula la profundidad del agua
water_depth = newton_raphson(initial_guess)

print("La profundidad del agua en el abrevadero es:", round(water_depth, 2), "cm")
```

Un objeto que cae verticalmente a través del aire está sujeto a una resistencia 
viscosa, así como a la fuerza de gravedad. Suponga que un objeto con masa 𝑚 cae 
desde una altura 𝒔𝟎 y que la altura del objeto después de 𝑡 segundos es
𝑠(𝑡) = 𝑠0 −(𝑚𝑔/𝑘)𝑡 +𝑚^2𝑔/𝑘^2(1 − 𝑒^(−𝑘𝑡/𝑚) )

donde 𝑔 = 9.81 𝒎/𝒔^𝟐 y 𝑘 representa el coeficiente de la resistencia del aire en 𝑵𝒔/𝒎. 
Suponga 𝒔𝟎 = 300 𝑚, 𝑚 = 0.25 𝑘𝑔 y 𝑘 = 0.1 𝑵𝒔/𝒎. Encuentre, dentro de 0.01 𝑠𝑒𝑔𝑢𝑛𝑑𝑜𝑠, 
el tiempo que tarda un cuarto de kg en golpear el piso.

```python
import math

# Parámetros dados
s0 = 300  # Altura inicial en metros
m = 0.25  # Masa del objeto en kg
g = 9.81  # Aceleración debido a la gravedad en m/s^2
k = 0.1   # Coeficiente de resistencia del aire en Ns/m

# Función para calcular s(t)
def calculate_s(t):
    return s0 - (m * g / k) * t + (m**2 * g / k**2) * (1 - math.exp(-k * t / m))

# Derivada de la función s(t)
def derivative_s(t):
    return -(m * g / k) + (m**2 * g / k**2) * (k / m) * math.exp(-k * t / m)

# Método de Newton-Raphson para encontrar la raíz
def newton_raphson(guess, tolerance=0.01):
    while True:
        t_next = guess - calculate_s(guess) / derivative_s(guess)
        if abs(t_next - guess) < tolerance:
            return t_next
        guess = t_next

# Aproximación inicial
initial_guess = 1

# Calcula el tiempo que tarda en golpear el piso
time_to_hit_ground = newton_raphson(initial_guess)

print("El tiempo que tarda en golpear el piso es:", round(time_to_hit_ground, 2), "segundos")
```

Use el teorema 2.1 para encontrar una cota para el número de iteraciones necesarias 
para lograr una aproximación con precisión de 𝟏𝟎^−𝟒 para la solución de 𝒙^𝟑 − 𝒙 −𝟏 = 𝟎 que se encuentra dentro del intervalo [1, 2]. Encuentre una aproximación para la raíz con este grado de precisión.


```python
def f(x):
    return x**3 - x - 1

def metodo_biseccion(a, b, tol, max_iteraciones):
    for i in range(1, max_iteraciones + 1):
        c = (a + b) / 2
        if abs(f(c)) < tol:
            return c, i
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2, max_iteraciones

# Definir los parámetros
a = 1
b = 2
tol = 1e-4
max_iteraciones = 100 

# LLamamos a la función de método de bisección
raiz, iteraciones = metodo_biseccion(a, b, tol, max_iteraciones)

# Mostrar el resultado
print("Aproximación de la raíz:", raiz)
print("Número de iteraciones:", iteraciones)
```
